```python
# Create a mask for rows where splitting by '_' doesn't yield exactly four parts
mask = data['Page'].str.split('_').apply(lambda x: len(x) != 4)

# Filter and display those rows
problem_rows = data[mask]
display(problem_rows)
```

```python
import pandas as pd

# This regex:
#   (?P<Title>.+)     => Captures everything (including underscores) as the Title (greedy)
#   _(?P<Project>[^_]+)  => Captures the Project field (no underscores)
#   _(?P<Access>[^_]+)   => Captures the Access field
#   _(?P<Agent>[^_]+)$   => Captures the Agent field at the end
pattern = r'^(?P<Title>.+)_(?P<Project>[^_]+)_(?P<Access>[^_]+)_(?P<Agent>[^_]+)$'

data[['Title', 'Project', 'Access', 'Agent']] = data['Page'].str.extract(pattern)
display(data)
```

```python
# Identify rows where the extraction failed (any NaN in the new columns)
problem_rows = data[data[['Title', 'Project', 'Access', 'Agent']].isnull().any(axis=1)]
print(problem_rows)
# drop page column and reorder columns to have the new columns first
data = data.drop(columns=['Title'])
data['Title'] = data['Page']
data = data.drop(columns=['Page'])
display(data)
```

```python
# Extract the country code from the Project column
data['CountryCode'] = data['Project'].str.extract(r'^(.*?)\.wikipedia\.org$', expand=False)

# Assign a unique numeric value to each country code (starting at 1)
data['Project_numeric'] = pd.factorize(data['CountryCode'])[0] + 1

data.drop(columns=['CountryCode'], inplace=True)
```

```python
# Factorize the 'Access' column to assign each unique string a numeric value starting at 1
data['Access_numeric'] = pd.factorize(data['Access'])[0] + 1

# Display the first few rows to verify the changes
display(data[['Access', 'Access_numeric']].head())

data = data.drop(columns=['Access'])
```

```python
# same for Agent
data['Agent_numeric'] = pd.factorize(data['Agent'])[0] + 1

# Display the first few rows to verify the changes
display(data[['Agent', 'Agent_numeric']].head())

data = data.drop(columns=['Agent'])
display(data)
```

```python
import numpy as np
import torch
import torch.nn as nn
from skorch import NeuralNetRegressor

# Define the ForecastLSTM model for single-timestep prediction.
class ForecastLSTM(nn.Module):
    def __init__(
        self,
        input_seq_len,
        n_other_features,
        hidden_size,
        num_layers,
        hidden_nodes,
        linear_layers,
        device="cuda",
        bidirectional=False,
        dropout=0.1,
    ):
        super(ForecastLSTM, self).__init__()
        
        self.num_layers = num_layers
        self.input_seq_len = input_seq_len
        self.hidden_size = hidden_size
        self.device = device
        self.n_other_feat = n_other_features
        self.num_directions = 2 if bidirectional else 1

        # LSTM to process the sequential (lag) features.
        self.lstm = nn.LSTM(
            input_size=1,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=bidirectional,
            dropout=dropout / 2,
        )

        # Encoding layer for the static features.
        self.encoding_layer = nn.Sequential(
            nn.Linear(self.n_other_feat, hidden_nodes),
            nn.ReLU(),
            nn.Linear(hidden_nodes, hidden_nodes),
        )

        # Build linear layers to combine LSTM output and encoded static features.
        # The input dimension equals LSTM output dimension plus the encoded static feature dimension.
        layers = []
        input_dim = hidden_size * self.num_directions + hidden_nodes
        for i in range(linear_layers - 1):
            layers.append(nn.Linear(input_dim, hidden_nodes))
            layers.append(nn.Dropout(dropout))
            layers.append(nn.ReLU())
            input_dim = hidden_nodes
        layers.append(nn.Linear(input_dim, 1))
        self.linear = nn.Sequential(*layers)

    def forward(self, x):
        # No need to force-move x, since skorch moves it automatically.
        # Instead, obtain the device from the module parameters:
        device = next(self.parameters()).device

        # Segment the input tensor into static and sequential parts.
        other_feat_x = x[:, :self.n_other_feat]             # shape: (batch, n_other_feat)
        input_seq_x = x[:, self.n_other_feat:]                # shape: (batch, input_seq_len)
        # Reshape sequential part for the LSTM: (batch, input_seq_len, 1)
        input_seq_x = input_seq_x.unsqueeze(-1)

        # Initialize LSTM hidden and cell states on the same device.
        hidden_states = torch.zeros(
            self.num_layers * self.num_directions,
            x.size(0),
            self.hidden_size,
            device=device,
        )
        cell_states = torch.zeros(
            self.num_layers * self.num_directions,
            x.size(0),
            self.hidden_size,
            device=device,
        )

        # Process sequential data with the LSTM.
        h_all, (h_out, _) = self.lstm(input_seq_x, (hidden_states, cell_states))
        
        # Use only the final hidden state for prediction.
        if self.num_directions == 1:
            lstm_out = h_all[:, -1, :]  # shape: (batch, hidden_size)
        else:
            # For bidirectional LSTM, concatenate final states from both directions.
            lstm_out = torch.cat((h_out[-2], h_out[-1]), dim=1)

        # Process static features through the encoding layer.
        encoded_other = self.encoding_layer(other_feat_x)  # shape: (batch, hidden_nodes)

        # Concatenate LSTM output and encoded static features.
        concat = torch.cat((lstm_out, encoded_other), dim=1)

        # Get final prediction.
        out = self.linear(concat).squeeze(-1)
        return out
```

```python
from skorch.helper import predefined_split
from skorch.callbacks import EarlyStopping
from skorch.dataset import Dataset
# Parameters.
n_other_features = 8    # Number of static features
input_seq_len = 7       # Number of lag features
print(f"X_test.shape: {X_test.shape}")
print(f"X_train.shape: {X_train.shape}")
print(f"y_test.shape: {y_test.shape}")
print(f"y_train.shape: {y_train.shape}")
# Assume X_train is a numpy array of shape (n_samples, n_other_features + input_seq_len)
# and y_train is a pandas Series or numpy array of shape (n_samples,)
# Ensure the data is of type float32.
X_train_np = X_train.astype(np.float32)
y_train_np = y_train.astype(np.float32)
# normalize the views with log1p

# cut the data to the first 20000 rows
X_train_np = X_train_np[:20000]
y_train_np = y_train_np[:20000]
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Assuming X_test and y_test are defined and represent your test set,
# convert them to NumPy arrays of type float32.
X_test_np = X_test.astype(np.float32)
y_test_np = y_test.astype(np.float32)
print(f"y_test_np.shape: {y_test_np.shape}")
# cut the data to the first 20000 rows
X_test_np = X_test_np[:20000]
y_test_np = y_test_np[:20000]

valid_dataset = Dataset(X_test_np, y_test_np)
```

```python
# %%
import torch
import torch.optim as optim
from skorch import NeuralNetRegressor
from skorch.callbacks import EarlyStopping
from skorch.helper import predefined_split
from ray import tune
import numpy as np

# Assume the following are defined in prior cells:
#   - X_train_np: numpy array of shape (n_samples_train, n_other_features + input_seq_len)
#   - y_train_np: numpy array of log1p-transformed targets of shape (n_samples_train,)
#   - valid_dataset: a skorch Dataset built from your X_test_np and y_test_np data
#   - input_seq_len: number of sequential (lag) features (e.g. 7)
#   - n_other_features: number of static features (e.g. 8)
#   - ForecastLSTM: your model class
device = 'cuda' if torch.cuda.is_available() else 'cpu'

def objective_skorch(config):
    # Instantiate the NeuralNetRegressor with fixed parameters and hyperparameters from config.
    net = NeuralNetRegressor(
        module=ForecastLSTM,
        module__input_seq_len=input_seq_len,        # fixed: number of lag features
        module__n_other_features=n_other_features,    # fixed: number of static features
        module__hidden_size=config["module__hidden_size"],
        module__num_layers=config["module__num_layers"],
        module__hidden_nodes=config["module__hidden_nodes"],
        module__linear_layers=config["module__linear_layers"],
        module__device=device,                        # pass the device for module initialization if used
        max_epochs=100,                               # maximum number of epochs (adjust as needed)
        lr=config["lr"],
        optimizer=optim.Adam,
        batch_size=config["batch_size"],
        iterator_train__shuffle=False,
        callbacks=[('early_stopping', EarlyStopping(patience=10, monitor='valid_loss'))],
        train_split=predefined_split(valid_dataset),  # use pre-defined validation set
        verbose=0,
        device=device,
        iterator_train__pin_memory=True,
    )
    
    # Fit the model using the training data.
    net.fit(X_train_np, y_train_np)
    
    # Retrieve the best (minimum) validation loss from the training history.
    valid_losses = np.array(net.history[:, 'valid_loss'])
    best_loss = float(valid_losses.min())
    
    # Report the best loss as a dictionary (this avoids the unexpected keyword argument error).
    tune.report({"mean_loss": best_loss})

# Define the search space for LSTM-related hyperparameters.
search_space = {
    "module__hidden_size": tune.choice([64, 128, 256]),
    "module__num_layers": tune.choice([1, 2, 3]),
    "module__hidden_nodes": tune.choice([32, 64, 128]),
    "module__linear_layers": tune.choice([1, 2]),
    "lr": tune.loguniform(1e-4, 1e-2),
    "batch_size": tune.choice([256 * 4, 256 * 8]),
}

# Run Ray Tune hyperparameter sweep.
analysis = tune.run(
    objective_skorch,
    config=search_space,
    resources_per_trial={"cpu": 2, "gpu": 0.5} if device == "cuda" else {"cpu": 2},
    metric="mean_loss",
    mode="min",
    num_samples=20,  # Increase as needed
    verbose=1
)

print("Best hyperparameters found were:", analysis.best_config)
```
