```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python

# Load the data
data = pd.read_csv('../data/train_1/train_1.csv')
display(data)
```

```python
# print the dimensions of the dataset
print(data.shape)
```

```python
# Create a mask for rows where splitting by '_' doesn't yield exactly four parts
mask = data['Page'].str.split('_').apply(lambda x: len(x) != 4)

# Filter and display those rows
problem_rows = data[mask]
display(problem_rows)
```

```python
import pandas as pd

# This regex:
#   (?P<Title>.+)     => Captures everything (including underscores) as the Title (greedy)
#   _(?P<Project>[^_]+)  => Captures the Project field (no underscores)
#   _(?P<Access>[^_]+)   => Captures the Access field
#   _(?P<Agent>[^_]+)$   => Captures the Agent field at the end
pattern = r'^(?P<Title>.+)_(?P<Project>[^_]+)_(?P<Access>[^_]+)_(?P<Agent>[^_]+)$'

data[['Title', 'Project', 'Access', 'Agent']] = data['Page'].str.extract(pattern)
display(data)
```

```python
# Identify rows where the extraction failed (any NaN in the new columns)
problem_rows = data[data[['Title', 'Project', 'Access', 'Agent']].isnull().any(axis=1)]
print(problem_rows)
# drop page column and reorder columns to have the new columns first
data = data.drop(columns=['Title'])
data['Title'] = data['Page']
data = data.drop(columns=['Page'])
display(data)
data.drop(['Access', 'Agent'], axis=1, inplace=True)
```

```python
# Extract the country code from the Project column
data['CountryCode'] = data['Project'].str.extract(r'^(.*?)\.wikipedia\.org$', expand=False)

# Assign a unique numeric value to each country code (starting at 1)
data['Project_numeric'] = pd.factorize(data['CountryCode'])[0] + 1

data.drop(columns=['CountryCode'], inplace=True)
```

```python
display(data[['Project' , 'Project_numeric']])
data = data.drop(columns=['Project'])
```

```python
# Identify the date columns using a regex filter (assuming they follow a YYYY-MM-DD pattern)
date_columns = data.columns[data.columns.str.match(r'\d{4}-\d{2}-\d{2}')]

# Group by 'Project_numeric' and sum the page view columns
df_merged = data.groupby('Project_numeric')[date_columns].sum().reset_index()

# Display the new data frame
display(df_merged)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert the date column names to datetime objects
date_columns = df_merged.columns[df_merged.columns.str.match(r'\d{4}-\d{2}-\d{2}')]
dates = pd.to_datetime(date_columns)

plt.figure(figsize=(12, 6))

# Loop over each row in the merged dataframe
for _, row in df_merged.iterrows():
    views = row[date_columns].values
    plt.plot(dates, views, label=f"Project {int(row['Project_numeric'])}")

plt.xlabel("Date")
plt.ylabel("Total Page Views")
plt.title("Page Views Over Time by Project Numeric")
plt.legend(title="Project Numeric")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Identify the date columns (assuming they follow the YYYY-MM-DD format)
date_columns = df_merged.columns[df_merged.columns.str.match(r'\d{4}-\d{2}-\d{2}')].tolist()

# Melt the dataframe to transform it into long format: one row per Date-Views combo
df_long = df_merged.melt(id_vars=['Project_numeric'], 
                         value_vars=date_columns, 
                         var_name='Date', 
                         value_name='Views')

# Convert Date column to datetime format
df_long['Date'] = pd.to_datetime(df_long['Date'])

# Optionally drop rows where Views are NaN
df_long.dropna(subset=['Views'], inplace=True)

# Reset index after melting
df_long.reset_index(drop=True, inplace=True)

```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Ensure 'Views' column is numeric
df_long['Views'] = pd.to_numeric(df_long['Views'], errors='coerce')

# Sort the data by 'Project_numeric' and 'Date'
df_long.sort_values(by=['Project_numeric', 'Date'], inplace=True)

# Create lag features for the previous 7 days per project
for i in range(1, 8):
    df_long[f'lag{i}'] = df_long.groupby('Project_numeric')['Views'].shift(i)

# Create additional date features
df_long['dayofweek'] = df_long['Date'].dt.dayofweek
df_long['dayofmonth'] = df_long['Date'].dt.day
df_long['dayofyear'] = df_long['Date'].dt.dayofyear
df_long['month'] = df_long['Date'].dt.month
df_long['year'] = df_long['Date'].dt.year

# Remove rows with NaN values in the lag features
lag_features = [f'lag{i}' for i in range(1, 8)]
df_long.dropna(subset=lag_features, inplace=True)
# create lead features for the next 6 days
for i in range(1, 7):
    df_long[f'lead{i}'] = df_long.groupby('Project_numeric')['Views'].shift(-i)
lead_features = [f'lead{i}' for i in range(1, 7)]
# Remove rows with NaN values in the lead features
df_long.dropna(subset=lead_features, inplace=True)
# Display the first few rows of the processed dataframe
display(df_long.head())

# Define train-test split date
split_date = '2016-06-01'
train = df_long[df_long['Date'] < split_date]
test = df_long[df_long['Date'] >= split_date]
raise
```

```python
del df_long
del data

# Select features and target
features = ['Project_numeric',  'lag1', 'lag2', 'lag3', 'lag4', 'lag5', 'lag6', 'lag7', 'dayofweek', 'dayofmonth', 'dayofyear', 'month', 'year']

target = ['Views', 'lead1', 'lead2', 'lead3', 'lead4', 'lead5', 'lead6']




X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]
# normalize X with min max scaler apparently normalizing y is problematic for 
# Xgboost and other tree based models
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


print("Training data shape:", X_train.shape, y_train.shape)
print("Testing data shape:", X_test.shape, y_test.shape)
```

```python
display(X_test[:5])
display(y_test[:5])
import xgboost as xgb

# Convert to DMatrix format (XGBoost's optimized data structure)
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
# clean up the data to free up memory

```

```python
# display shape of dtrain and dtest
print(dtrain.num_row(), dtrain.num_col())
print(dtest.num_row(), dtest.num_col())
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'max_depth': 10,  # Increase depth to capture more patterns
    'eta': 0.09,  # Lower learning rate for better convergence
    'subsample': 0.8,  
    'colsample_bytree': 0.8,  
    'lambda': 1.0,  # L2 regularization to control overfitting
    'alpha': 0.1,  # L1 regularization for feature selection
    'device': 'cuda', # Use GPU acceleration
    'tree_method': 'hist' # put on gpu if available
}

# perform grid search to find the best parameters
# Grid search is computationally expensive, so it's best to use a subset of the data
# Define the parameter grid
param_grid = {
    'max_depth': [5, 10, 15],
    'eta': [0.01, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'lambda': [0.1, 1.0, 10.0],
    'alpha': [0.1, 1.0, 10.0]
}
# Create a dictionary to store the best parameters
best_params = {}
# Create a fixed set of parameters
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'device': 'gpu',
    'tree_method': 'gpu_hist'
}
# use gridsearchcv to find the best parameters
from sklearn.model_selection import GridSearchCV
# Create an XGBoost regressor
model = xgb.XGBRegressor()
# Create a GridSearchCV object
grid = GridSearchCV(model, param_grid, n_jobs=5, cv=3)
# Fit the GridSearchCV object
# create a subset of the data to speed up the grid search
X_train_subset = X_train[:10000]
y_train_subset = y_train[:10000]

grid.fit(X_train_subset, y_train_subset)
# Get the best parameters
best_params = grid.best_params_
print(best_params)
# Update the parameters with the best ones
params.update(best_params)
```

```python
# Train with early stopping
num_round = 10000  # Increase total rounds, but use early stopping
model = xgb.train(
    params, dtrain, num_round, evals=[(dtest, 'test')],
    early_stopping_rounds=50,  # Stops if no improvement in 50 rounds
    verbose_eval=10
)
```

```python
# get training L2 loss
train_rmse = model.eval(dtrain)
test_rmse = model.eval(dtest)
print("Training RMSE:", train_rmse)
print("Testing RMSE:", test_rmse)

# get L1 loss
train_mae = model.eval(dtrain, 'mae')
test_mae = model.eval(dtest, 'mae')
print("Training MAE:", train_mae)
print("Testing MAE:", test_mae)


features = ['Project_numeric', 'lag1', 'lag2', 'lag3', 'lag4', 'lag5', 'lag6', 'lag7', 'dayofweek', 'dayofmonth', 'dayofyear', 'month', 'year']
# Directly assign feature names to the Booster object
model.feature_names = features

# Plot the most important features with updated y-axis labels
xgb.plot_importance(model)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import xgboost as xgb

# Define target labels corresponding to the forecast horizon:
# Day 0 is the current "Views" value, followed by lead1, lead2, ..., lead6.
target_labels = ['Views', 'lead1', 'lead2', 'lead3', 'lead4', 'lead5', 'lead6']

# Choose 7 random indices from the test DataFrame
random_indices = random.sample(test.index.tolist(), 1)
forecast_horizon = list(range(len(target_labels)))  # [0, 1, 2, 3, 4, 5, 6]

# Determine grid dimensions for subplots (e.g., 3 columns)
n_plots = len(random_indices)
ncols = 1
nrows = 1

fig, axes = plt.subplots(nrows, ncols, figsize=(15, nrows * 4), sharex=True)

for i, idx in enumerate(random_indices):
    # Select a random test sample
    sample_row = test.loc[idx]
    sample_date = sample_row['Date']
    # Actual forecast series as a list of floats
    actual_series = sample_row[target_labels].astype(float).values.tolist()
    
    # Extract features, scale them, and create a DMatrix for prediction
    sample_features = sample_row[features]
    sample_features_scaled = scaler.transform(sample_features.values.reshape(1, -1))
    sample_dmatrix = xgb.DMatrix(sample_features_scaled, feature_names=features)
    
    # Get predictions and ensure they are floats
    predicted_series = model.predict(sample_dmatrix)[0]
    predicted_series = [float(val) for val in predicted_series]
    
    ax = axes
    # Plot actual and predicted series over the forecast horizon
    ax.plot(forecast_horizon, actual_series, marker='o', label='Actual')
    ax.plot(forecast_horizon, predicted_series, marker='x', label='Predicted')
    ax.set_title(f"Forecast starting {sample_date.strftime('%Y-%m-%d')}")
    ax.set_xlabel("Forecast Day (0 = current)")
    ax.set_ylabel("Page Views")
    ax.legend()

# Remove any unused subplots if there are extra axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
