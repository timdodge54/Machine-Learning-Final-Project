```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python

# Load the data
data = pd.read_csv('./data/train_1/train_1.csv')
display(data)
```

```python
# print the dimensions of the dataset
print(data.shape)
```

```python
# Create a mask for rows where splitting by '_' doesn't yield exactly four parts
mask = data['Page'].str.split('_').apply(lambda x: len(x) != 4)

# Filter and display those rows
problem_rows = data[mask]
display(problem_rows)
```

```python
import pandas as pd

# This regex:
#   (?P<Title>.+)     => Captures everything (including underscores) as the Title (greedy)
#   _(?P<Project>[^_]+)  => Captures the Project field (no underscores)
#   _(?P<Access>[^_]+)   => Captures the Access field
#   _(?P<Agent>[^_]+)$   => Captures the Agent field at the end
pattern = r'^(?P<Title>.+)_(?P<Project>[^_]+)_(?P<Access>[^_]+)_(?P<Agent>[^_]+)$'

data[['Title', 'Project', 'Access', 'Agent']] = data['Page'].str.extract(pattern)
display(data)
```

```python
# Identify rows where the extraction failed (any NaN in the new columns)
problem_rows = data[data[['Title', 'Project', 'Access', 'Agent']].isnull().any(axis=1)]
print(problem_rows)
# drop page column and reorder columns to have the new columns first
data = data.drop(columns=['Title'])
data['Title'] = data['Page']
data = data.drop(columns=['Page'])
display(data)
```

```python
# Extract the country code from the Project column
data['CountryCode'] = data['Project'].str.extract(r'^(.*?)\.wikipedia\.org$', expand=False)

# Assign a unique numeric value to each country code (starting at 1)
data['Project_numeric'] = pd.factorize(data['CountryCode'])[0] + 1

data.drop(columns=['CountryCode'], inplace=True)
```

```python
display(data[['Project' , 'Project_numeric']])
data = data.drop(columns=['Project'])
```

```python
# Factorize the 'Access' column to assign each unique string a numeric value starting at 1
data['Access_numeric'] = pd.factorize(data['Access'])[0] + 1

# Display the first few rows to verify the changes
display(data[['Access', 'Access_numeric']].head())

data = data.drop(columns=['Access'])
```

```python
# same for Agent
data['Agent_numeric'] = pd.factorize(data['Agent'])[0] + 1

# Display the first few rows to verify the changes
display(data[['Agent', 'Agent_numeric']].head())

data = data.drop(columns=['Agent'])
display(data)
```

```python
import pandas as pd

# Define the date columns 
date_columns = data.columns[:-4]  # Assuming last 4 columns are Title, Project, Project_numeric, etc.

# Melt the dataframe to transform it into long format i.e. one row per Date-Pageview combo
df_long = data.melt(id_vars=['Title', 'Project_numeric', 'Access_numeric', 'Agent_numeric'],
                  value_vars=date_columns, var_name='Date', value_name='Views')  # Renamed 'Pageviews' to 'Views'

# Convert Date column to datetime format
df_long['Date'] = pd.to_datetime(df_long['Date'])

# Drop any rows where Views are NaN (optional, depending on your needs)
df_long.dropna(subset=['Views'], inplace=True)

# Reset index after melting
df_long.reset_index(drop=True, inplace=True)

# Display the first few rows
print(df_long.head())
```

```python
# import scalar so we can normalize the data
from sklearn.preprocessing import MinMaxScaler
# Ensure 'Views' column is numeric
df_long['Views'] = pd.to_numeric(df_long['Views'], errors='coerce')

# Sort the data by Date
df_long.sort_values(by=['Date'], inplace=True)

# Create lag features (This is create a sequence of lagged values for each Title)
# This is essentially turning the time series data into a supervised learning problem
df_long['lag1'] = df_long.groupby('Title')['Views'].shift(1)
df_long['lag2'] = df_long.groupby('Title')['Views'].shift(2)
df_long['lag3'] = df_long.groupby('Title')['Views'].shift(3)
df_long['lag4'] = df_long.groupby('Title')['Views'].shift(4)
df_long['lag5'] = df_long.groupby('Title')['Views'].shift(5)
df_long['lag6'] = df_long.groupby('Title')['Views'].shift(6)
df_long['lag7'] = df_long.groupby('Title')['Views'].shift(7)

# create date features
df_long['dayofweek'] = df_long['Date'].dt.dayofweek
df_long['dayofmonth'] = df_long['Date'].dt.day
df_long['dayofyear'] = df_long['Date'].dt.dayofyear
df_long['month'] = df_long['Date'].dt.month
df_long['year'] = df_long['Date'].dt.year


# remove rows with NaN values only want to predict on rows where we have all the data 
# (we can deal with missing values in the future)
df_long.dropna(subset=['lag1', 'lag2', 'lag3', 'lag4', 'lag5', 'lag6', 'lag7'], inplace=True)
display(df_long.head())

# Define train-test split date
split_date = '2016-06-01'
train = df_long[df_long['Date'] < split_date]
test = df_long[df_long['Date'] >= split_date]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot the distribution of Views before scaling
plt.hist(df_long['Views'], bins=50)
plt.title('Distribution of Views (Before Scaling)')
plt.show()

# Apply log transformation if the distribution is skewed
df_long['log_views'] = np.log1p(df_long['Views'])
```

```python
del df_long
del data

# Select features and target
features = ['Project_numeric', 'Access_numeric', 'Agent_numeric', 'lag1', 'lag2', 'lag3', 'lag4', 'lag5', 'lag6', 'lag7', 'dayofweek', 'dayofmonth', 'dayofyear', 'month', 'year']

target = 'Views'  




X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]
# normalize X with min max scaler apparently normalizing y is problematic for 
# Xgboost and other tree based models
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


print("Training data shape:", X_train.shape, y_train.shape)
print("Testing data shape:", X_test.shape, y_test.shape)
```

```python
display(X_test[:5])
display(y_test[:5])
import xgboost as xgb

# Convert to DMatrix format (XGBoost's optimized data structure)
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
# clean up the data to free up memory

```

```python
# display shape of dtrain and dtest
print(dtrain.num_row(), dtrain.num_col())
print(dtest.num_row(), dtest.num_col())
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'max_depth': 10,  # Increase depth to capture more patterns
    'eta': 0.09,  # Lower learning rate for better convergence
    'subsample': 0.8,  
    'colsample_bytree': 0.8,  
    'lambda': 1.0,  # L2 regularization to control overfitting
    'alpha': 0.1,  # L1 regularization for feature selection
    'device': 'cuda', # Use GPU acceleration
    'tree_method': 'hist' # put on gpu if available
}

# perform grid search to find the best parameters
# Grid search is computationally expensive, so it's best to use a subset of the data
# Define the parameter grid
param_grid = {
    'max_depth': [5, 10, 15],
    'eta': [0.01, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'lambda': [0.1, 1.0, 10.0],
    'alpha': [0.1, 1.0, 10.0]
}
# Create a dictionary to store the best parameters
best_params = {}
# Create a fixed set of parameters
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'device': 'gpu',
    'tree_method': 'gpu_hist'
}
# use gridsearchcv to find the best parameters
from sklearn.model_selection import GridSearchCV
# Create an XGBoost regressor
model = xgb.XGBRegressor()
# Create a GridSearchCV object
grid = GridSearchCV(model, param_grid, n_jobs=5, cv=3)
# Fit the GridSearchCV object
# create a subset of the data to speed up the grid search
X_train_subset = X_train[:10000]
y_train_subset = y_train[:10000]

grid.fit(X_train_subset, y_train_subset)
# Get the best parameters
best_params = grid.best_params_
print(best_params)
# Update the parameters with the best ones
params.update(best_params)
```

```python
# Train with early stopping
num_round = 10000  # Increase total rounds, but use early stopping
model = xgb.train(
    params, dtrain, num_round, evals=[(dtest, 'test')],
    early_stopping_rounds=50,  # Stops if no improvement in 50 rounds
    verbose_eval=10
)
```

```python
# get training L2 loss
train_rmse = model.eval(dtrain)
test_rmse = model.eval(dtest)
print("Training RMSE:", train_rmse)
print("Testing RMSE:", test_rmse)

# get L1 loss
train_mae = model.eval(dtrain, 'mae')
test_mae = model.eval(dtest, 'mae')
print("Training MAE:", train_mae)
print("Testing MAE:", test_mae)


# Plot the most important features
xgb.plot_importance(model)
plt.show()
# plot a series of time truth views vs predicted views for a single article

# get a valid article
article = test['Title'].iloc[0]
# filter the data   
article_data = test[test['Title'] == article]
# generate predictions for the article
# segment into features and target
features = ['Project_numeric', 'Access_numeric', 'Agent_numeric', 'lag1', 'lag2', 'lag3', 'lag4', 'lag5', 'lag6', 'lag7', 'dayofweek', 'dayofmonth', 'dayofyear', 'month', 'year']
target = 'Views'
# create the DMatrix
article_data['Prediction'] = model.predict(xgb.DMatrix(scaler.transform(article_data[features])))
# plot the results
plt.figure(figsize=(15, 5))
plt.plot(article_data['Date'], article_data['Views'], label='True Views')
plt.plot(article_data['Date'], article_data['Prediction'], label='Predicted Views')
plt.title(article)
plt.legend()
plt.show()
```
